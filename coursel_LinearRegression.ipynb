```python
# import house price
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
print(data.shape)
data.head()
```

数据预处理：

数太大，归一化，让数据的分布处于同一区间，我们选择一种简单的数据调整方法，每个数除以其最大值。

```python
# preprocessing the data
data_select = data[['BedroomAbvGr', 'LotArea', 'Neighborhood', 'SalePrice']]
data_select = data_select.rename(columns = {'BedroomAbvGr' : 'room', 'LotArea' : 'area'})
data_select = data_select.dropna(axis=0) # drop nan value
for col in np.take(data_select.columns, [0,1, -1]):
    data_select[col] /= data_select[col].max()
data_select.head()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data_select.copy(), test_size = 0.9)
print(train.shape, test.shape)
```

### 2.2 构建模型

#### 2.2.1 线性回归模型

从最简单的情况出发，构建一台预测价格的机器，假设函数 h(x) = wx + b 是线性的

```python
def linear(features, pars): # pars include w and b
    price = np.sum(features*pars[:-1], axis = 1) + pars[-1]
    return price
```

#### 2.2.2 模型测试

试一试你的模型预测价格，和真实价格做比较

```python
# par1 = 0.1， par2 = 0.1
train['predict'] = linear(train[['room', 'area']].values, np.array([0.1, 0.1, 0.0]))
train.head()
```

 可以看到，在该参数下，模型的预测价格和真实价格有较大的差距。那么寻找合适的参数值是我们该做的事情。

#### 2.2.3 代价函数

预测函数为 h(x) = wx + b

度量预测错误， 代价函数是误差平方和。

$$J(w) = \frac{1}{2}\sum_{i=1}^{m}{(h(x^i)-y^i)^2}$$

$$ w* = argmain  J(x) = argmin  \frac{1}{2}\sum_{i=1}^{m}{(h(x^i)-y^i)^2}$$

```python
def mean_squared_error(y_pred, y):
    return sum(np.array(y_pred - y)**2)

def Cost(df, features, pars):
    df['predict'] = linear(df[features].values, pars)
    cost = mean_squared_error(df.predict, df.SalePrice)/len(df)
    return cost

cost = Cost(train, ['room', 'area'], np.array([0.1, 0.1, 0.0]))
print(cost)
```

#### 2.2.4 暴力算法

我们需要找到一组参数，线性回归模型代价最少！

最简单的办法就是用 格点搜索法。 为了减少演演算时间，假设模型的截距项 b=0

```python
Xs = np.linspace(0,1,100)
Ys = np.linspace(0,1,100)
Zs = np.zeros([100, 100])

Xs, Ys = np.meshgrid(Xs, Ys)
Xs.shape, Ys.shape
```

```python
w1 = []
w2 = []
costs = []

for i in range(100):
    for j in range(100):
        w1.append(0.01*i)
        w2.append(0.01*j)
        costs.append(Cost(train, ['room','area'], np.array([0.01*i, 0.01*j, 0])))
        
# argmin(): Returns the indices of the minimum values along an axis
index = np.array(costs).argmin()
```

#### 2.2.5 梯度下降

```python
# pars includes w and b
def gradient(train, features, pars):
    Gradient = np.zeros(len(pars))
    for i in range(len(pars)):
        pars_new = pars.copy()
        pars_new[i] += 0.01
        Gradient[i]  = (Cost(train, features, pars_new) - Cost(train, features, pars))/0.01
    return Gradient

gradient(train, ['room', 'area'], [0.2, 0.1, 0])
```

```python
def GradientDescent(data, epochs, lr, features, pars):
    Costs = []
    for i in range (epochs):
        grad = gradient(data, features, pars)
        if i%50 == 0:
            Costs.append(Cost(data, features, pars))
        pars -= grad*lr
    print('w = ', pars)
    return pars, Costs
```

```python
pars, costs_grad = GradientDescent(train, 500, 0.002, ['room', 'area'], [0.1,0.1, 0])
print( pars, costs_grad)
```

```python
# plot the gradient cost graph

plt.plot(costs_grad)
plt.xlabel('time')
plt.ylabel('cost')
```

### 2.4 如何做得更好

结果是否理想，能够如何优化，是否存在一个模型， 使得误差为0？ 增加的特征试一试！
我们把之前的模型封装起来，并且使用公式运算梯度，比较一下和之前的速度差别！

```python
class LinearRegressionGD(object):
    
    def __init__(self, eta=0.001, n_iter=20):
        self.eta = eta
        self.n_iter = n_iter
        
    def fit(self, X, y):
        self.w_ = np.zeros(1+X.shape[1])
        self.cost_ = []
        
        for i in range(self.n_iter):
            output = self.net_input(X)
            errors = (y - output)
            self.w_[:-1] += self.eta * X.T.dot(errors)
            self.w_[-1] += self.eta * errors.sum()
            cost = (errors**2).sum() /2.0
            self.cost_.append(cost)
        return self
    
    def net_input(self, X):
        return np.dot(X, self.w_[:-1] + self.w_[-1])
    
    def predict(self, X):
        return self.net_input(X)
        
```

```python
data_select = pd.get_dummies(data_select) # convert non-digital features to numbers eg: areas
train, test = train_test_split(data_select.copy(), test_size=0.9)
```

```python
trian_ = train.copy()
trian_y = trian_.pop('SalePrice')
train_x = trian_

test_ = test.copy()
test_y = test_.pop('SalePrice')
test_x = test_
```

```python
lr = LinearRegressionGD(n_iter = 1000)
lr.fit(train_x.values, trian_y.values)
```

```python
test['predict'] = lr.predict(test_x.values)
print('MSE test: %.3f' % (mean_squared_error(test_y, test['predict'])))
test.head()
```
